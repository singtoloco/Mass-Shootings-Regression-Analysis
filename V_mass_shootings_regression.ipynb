```python
import pandas as pd
import matplotlib.pyplot as plt
# import gmaps
# import gmaps.geojson_geometries
# import gmaps.datasets

import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

import seaborn as sns
```

## ===================

```python
#using with dummies (Red/Blue States) 

combined_all = pd.read_csv('OutputFiles/Tina/linear_with_dummies.csv')

combined_all['Gun Check per capita'] = combined_all['Gun Check Totals']/combined_all['Population']
# combined_all.loc[:,'Gun Check per capita'] *= 10**5
```

```python
## 'Per Capita Hate Group' was calculated with multiplying with 100,000

# combined_all.loc[:,'Per Capita Hate Group'] *= 10
```

```python
print(combined_all.shape)
```

```python
## Dropping 'District of Columbia' b/c of special cases and potentially misleading in Regression

combined_all = combined_all.loc[combined_all['States'] != 'District of Columbia']
```

```python
print(combined_all.shape)
```

### Start regression on combined_all

```python
x_data = combined_all[['Poverty Rates', 'Per Capita Hate Group', 'GRADE number', 'Gun Check per capita', 'Election_Result_B', 'Election_Result_R']]

y_data = combined_all['Per Capita Mass Shooting']

##Adding constant term to the linear regression model
x_data = sm.add_constant(x_data)

model = sm.OLS(y_data, x_data).fit()
predictions = model.predict(x_data)

print(model.summary())
```

```python
# Basic correlogram

linear_corrplot = combined_all[['Per Capita Mass Shooting', 'Poverty Rates', 'Per Capita Hate Group', 'GRADE number', 'Gun Check per capita', 'Election_Result_B', 'Election_Result_R']]
sns.pairplot(linear_corrplot)
plt.show()
# plt.savefig('Images/Vanich/corr_Regression_1.png')
```
